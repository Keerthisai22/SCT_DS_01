```python
#histogram
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("population_data.csv")
plt.figure(figsize=(5,2))
plt.hist(df["Age"],bins=[0,20,40,60,80,100],color='pink',edgecolor='black',rwidth=0.8)
plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Number of people")
plt.show()
#bar chart
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("population_data.csv")
plt.figure(figsize=(5,2))
gender_counts=df["Gender"].value_counts()
gender_counts.plot(kind='bar',color=['pink','red','green'],edgecolor='black')
plt.title("gender distribution")
plt.xlabel("Gender")
plt.ylabel("Number of people")
plt.show()
```
